Importing important libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

Getting Data via given Link

```python
link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(link) 
soup = BeautifulSoup(page.text)
```

```python
table = soup.find('table')
```

```python
ta_ro = table.find_all('tr')
result = []
for tr in ta_ro:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        result.append(row)
df = pd.DataFrame(result, columns=['PostalCode', 'Borough', 'Neighbourhood'])
df.head()
```

Dropping those rows which don't have any value or Not Assigned

```python
df=df.drop(df[df.Borough=="Not assigned"].index)
df
```

Copying value of Borough to Neighbourhood if value is not assigned

```python
for index, row in df.iterrows():
    if row['Neighbourhood']=="Not assigned":
        row['Neighbourhood']=row['Borough']
df.head()
```

Grouping rows which have same postal code and borough value

```python
df=df.groupby(['PostalCode', 'Borough'])['Neighbourhood'].apply(','.join).reset_index()
df
```

Replacing Neighbourhood Not assigned value with queen park

```python
df['Neighbourhood'].replace("Not assigned", "queens's park", inplace=True)
df
```

Getting longitude and latitude using Geo-spatial data(using given link a XML File) 

```python
df_geo_data=pd.read_csv('http://cocl.us/Geospatial_data')
df_geo_data.columns=['PostalCode', 'Latitude', 'Longitude']
```

We have existing dataframe df with coulumns Postalcode, Borough, Neighbourhood and a new dataframe df_geo_data having postal code, Longtude and Latitude

Joining the two Dataframes

```python
df_join=pd.merge(df, df_geo_data, on=['PostalCode'], how='inner')
df_latitude_longitude=df_join[['PostalCode', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude']].copy()
df_latitude_longitude
```
